```python
import pandas as pd
```

```python
df = pd.read_csv('../amz_uk_price_prediction_dataset.csv')
```

```python
df.drop("uid", inplace=True, axis=1)
```

```python
df
```

```python
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result
```

```python
crosstab_result['BestSellerProportion'] = crosstab_result[True] / (crosstab_result[True] + crosstab_result[False])
sorted_categories = crosstab_result.sort_values(by='BestSellerProportion', ascending=False)
print(sorted_categories)
```

```python
# The categories where the best-seller is more prevalent are Grocery, Smart Home Security & Lighting, Health & Personal Care
```

```python
from scipy.stats import chi2_contingency

chi2_statistic, chip2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chip2_p_value
```

```python
# The null hypothesis of the Chi-sqared test is that the two categorical variables are independent. 
# Given the low (p)-value, we reject the null hypothesis. 
# This implies that there's a significant association between category and isBestSeller in the dataset
```

```python
import numpy as np

num_rows = crosstab_result.shape[0]
num_cols = crosstab_result.shape[1]
n = crosstab_result.values.sum()
V = np.sqrt((chi2_statistic/n) / (min(num_rows, num_cols)-1))

print("Cramér's V:", V)
```

```python
# The Cramér's V value of 0.0866 suggests a weak association between the two variables. 
# Though statistically significant, the strength of this relationship is not very strong in practical terms
```

```python
import matplotlib.pyplot as plt

crosstab_result.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Best-Seller Status by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.legend(title='Best-Seller', labels=['Not Best-Seller', 'Best-Seller'])
plt.xticks(rotation=90, ha='right')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

```python
sorted_categories = crosstab_result.sort_values(by=True, ascending=False)

top_20_categories = sorted_categories.head(20)

top_20_categories.plot(kind='bar', stacked=True, figsize=(14, 14))
plt.title('Top 20 Best-Seller Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.legend(title='Best-Seller', labels=['Not Best-Seller', 'Best-Seller'])
plt.xticks(rotation=45, ha='right')
plt.yscale('log') 
plt.tight_layout()
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
df_no_outliers
```

```python
import seaborn as sns

top_20_categories = df_no_outliers['category'].value_counts().head(20).index
df_top_20_sample = df_no_outliers[df_no_outliers['category'].isin(top_20_categories)].sample(n=1000)

plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top_20_sample, x='category', y='price', palette='coolwarm', inner='quartile')
plt.title('Violin Plot of Product Prices for Top 20 Categories (Sample)')
plt.xlabel('Product Category')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_10_categories = df_no_outliers['category'].value_counts().head(10).index

df_top_10 = df_no_outliers[df_no_outliers['category'].isin(top_10_categories)]

average_prices = df_top_10.groupby('category')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=average_prices.index, y=average_prices.values, palette='coolwarm')
plt.title('Average Price of Products for Top 10 Categories')
plt.xlabel('Product Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# The product category with higher average price is Sports & Outdoors
```

```python
top_10_categories = df['category'].value_counts().head(10).index

df_top_10 = df[df['category'].isin(top_10_categories)]

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_top_10, x='category', y='stars', palette='coolwarm')
plt.title('Distribution of Product Ratings for Top 10 Categories')
plt.xlabel('Product Category')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
